```python
import pandas as pd

# Load appearances.csv
appearances = pd.read_csv('appearances.csv')

# 1. Create Appearances Table
appearances_table = appearances[['gameID', 'playerID', 'goals', 'ownGoals', 'shots', 'assists', 'keyPasses', 'yellowCard', 'redCard', 'time', 'substituteIn', 'substituteOut']]

# 2. Create PlayerPosition Table
player_position_table = appearances[['playerID', 'position', 'positionOrder']].drop_duplicates()

# Save them
appearances_table.to_csv('final_appearances.csv', index=False)
player_position_table.to_csv('final_playerposition.csv', index=False)
```

```python
# Load teamstats.csv
teamstats = pd.read_csv('teamstats.csv')

# 1. Create TeamLocations Table
team_locations = teamstats[['gameID', 'location', 'teamID']]

# 2. Create TeamStats Table
team_stats = teamstats[['gameID', 'teamID', 'goals', 'shots', 'shotsOnTarget', 'fouls', 'corners', 'yellowCards', 'redCards', 'result']]

# Save them
team_locations.to_csv('final_teamlocations.csv', index=False)
team_stats.to_csv('final_teamstats.csv', index=False)
```
